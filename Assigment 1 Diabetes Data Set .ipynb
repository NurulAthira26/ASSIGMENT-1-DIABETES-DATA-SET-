```python

import pandas as pd
import numpy as np
from scipy import stats

# Assuming df is your DataFrame
df = pd.read_csv('diabetes_binary_5050split_health_indicators_BRFSS2015.csv')  # replace 'your_file.csv' with your actual file

# Flatten the DataFrame to a single list of values
values = df.values.flatten()

# Calculate mean
mean = np.mean(values)

# Calculate median
median = np.median(values)

# Calculate mode
mode = stats.mode(values)[0]

# Calculate standard deviation
std_dev = np.std(values)

# Calculate max
max_val = np.max(values)

# Calculate min
min_val = np.min(values)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Standard Deviation: {std_dev}")
print(f"Max: {max_val}")
print(f"Min: {min_val}")
```
